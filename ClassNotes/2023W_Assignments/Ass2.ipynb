```python
# Get user input
fv = float(input('Amount to save at age 60: $'))
while fv < 0:
    print('WTF?')
    fv = float(input('Amount to save at age 60: $'))

age = int(input('Age in years to start saving: '))
while age < 0:
    print('OK')
    age = int(input('Age in years to start saving: '))

rate = float(input('interest rate assumption: '))
while rate < 0:
    print('OK')
    rate = float(input('interest rate assumption: '))

rate = rate / 100 / 12

# Calculate monthly savings needed
n = (60 - age) * 12

pv = fv / (1 + rate) ** n
ms = (pv * rate) / (1 - (1 + rate) ** (-n))

print(f'To retire comfortably at age 60 with ${fv} ' 
      f'you should save ${ms} per month starting at age {age} '
      f'with an assumed rate of {rate*12*100}')

```

```python
filename = input('Enter the filename to be created: ')
if not filename[-4:] != '.csv':
    filename += '.csv'

f = open(filename, 'w')
f.write('Month,Sales ($)\n')

while True:
    month = input('Enter a month: ')
    sales = float(input('Enter a sales: '))
    while sales < 0:
        print('cannot be negative')
        sales = float(input('Enter a sales: '))
    
    f.write(f'{month},{sales}\n')

    q = input('要結束嗎? (y/n)')
    if q.lower() == 'y':
        break

f.close()
```
