```python
import pandas as pd

# Load datasets
nfl_games = pd.read_csv('nflgames.csv')
nfl_teams = pd.read_csv('nflteams.csv')
nfl_players = pd.read_csv('nflplayers.csv')
nfl_positions = pd.read_csv('nflpositions.csv')
```

```python
nfl_games['FieldType'] = nfl_games['FieldType'].apply(lambda type: type.strip())
nfl_games.head()
```

```python
home_df = nfl_games[['HomeTeamID', 'HomeScore']]
home_df = home_df.rename(columns={'HomeTeamID': 'TeamID', 'HomeScore': 'Score'})
home_df.head()
```

```python
away_df = nfl_games[['AwayTeamID', 'AwayScore']]
away_df = away_df.rename(columns={'AwayTeamID': 'TeamID', 'AwayScore': 'Score'})
away_df.head()
```

```python
game_df = pd.concat([home_df, away_df])
game_df
```

```python
team_avg = game_df.groupby('TeamID')['Score'].mean().reset_index()
max_avg = team_avg[team_avg['Score'] == team_avg['Score'].max()]
max_avg
```

2. Add Last Name Column and Identify Common Last Names in NFL Players

```python
# Extract last names
nfl_players['LastName'] = nfl_players['PlayerName'].apply(lambda name: name.split()[-1])

# Identify common last names
common_last_names = nfl_players['LastName'].value_counts()

print("Common last names:", common_last_names)
```

3. Create Pivot Table for Average HomeScore and AwayScore by FieldType

```python
# Pivot table
pivot_table = nfl_games.pivot_table(index='FieldType', values=['HomeScore', 'AwayScore'], aggfunc='mean')

print("Pivot table of average HomeScore and AwayScore by FieldType:\n", pivot_table)
```

4. Calculate Average Number of Players per Team in Each Position

```python
nfl_players.groupby(['TeamID', 'PositionID']).count().groupby('PositionID').mean().reset_index()
```

```python
# Merge datasets
merged_players = nfl_players.merge(nfl_positions, on='PositionID')

# Calculate average number of players per team in each position
merged_players.groupby(['TeamID', 'PositionName']).size().groupby('PositionName').mean()
```

```python
df = merged_players.groupby(['TeamID', 'PositionName']).size().groupby('PositionName').mean().reset_index()
df
```
